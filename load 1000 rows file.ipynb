```python
#with open(setting_file, 'rb') as f:
#    deduper = dedupe.StaticDedupe(f)
fields = [
            {'field': 'name_01', 'type': 'String'},
            {'field': 'street_01a', 'type': 'String'},
            {'field': 'street_01b', 'type': 'String', 'has missing': True},
            {'field': 'zip_01', 'type': 'Exact', 'has missing': True},
            {'field': 'city_01', 'type': 'String', 'has missing': True},
            {'field': 'nation_01', 'type': 'String', 'has missing': True},
            ]
```

```python
def preProcess(column):
#
    #column = unidecode(column)
    column = re.sub('  +', ' ', column)
    column = re.sub('\n', ' ', column)
    column = column.strip().strip('"').strip("'").lower().strip()
#If data is missing, indicate that by setting the value to None

    if not column:
        column = None
    return column

def readData(filename):
#
    data_d = {}
    with open(filename) as f:
        reader = csv.DictReader(f)
        for row in reader:
            clean_row = [(k, preProcess(v)) for (k, v) in row.items()]
            row_id = int(row['Id'])
            data_d[row_id] = dict(clean_row)

    return data_d
```

```python

data_d = readData(file)
```

```python
deduper = dedupe.Dedupe(fields)
deduper.prepare_training(data_d)
```

```python
 print('starting active labeling...')

dedupe.console_label(deduper)
```

```python
deduper.train()
```

```python
with open(training_file, 'w') as tf:
    deduper.write_training(tf)
```

```python
with open(settings_file, 'wb') as sf:
    deduper.write_settings(sf)
```

```python
print('clustering...')
clustered_dupes = deduper.partition(data_d, 0.5)

print('# duplicate sets', len(clustered_dupes))
```

```python
output_file = '/media/SharedSpace/temp_data_ops/data/Customers/dedup_Airbus.csv'
input_file = file

cluster_membership = {}
for cluster_id, (records, scores) in enumerate(clustered_dupes):
    for record_id, score in zip(records, scores):
        cluster_membership[record_id] = {
            "Cluster ID": cluster_id,
            "confidence_score": score
        }

with open(output_file, 'w') as f_output, open(input_file) as f_input:

    reader = csv.DictReader(f_input)
    fieldnames = ['Cluster ID', 'confidence_score'] + reader.fieldnames

    writer = csv.DictWriter(f_output, fieldnames=fieldnames)
    writer.writeheader()

    for row in reader:
        row_id = int(row['Id'])
        row.update(cluster_membership[row_id])
        writer.writerow(row)
```
